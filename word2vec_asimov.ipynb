```python
words = ['madre', 'padre', 'hombre', 'mujer', 'familia', 'oscuridad', 'luz', 'amigo', 'amigos',
         'hijo', 'hija', 'número', 'bien', 'juego', 'mal',
         'ataque', 'guerra', 'anciano', 'planeta', 'joven', 'viejo', 'edad', 'año', 'años',
         'día', 'días', 'tiempo', 'agua', 'estrella', 'vida', 'sol', 'eternidad', 'noche', 'nieto']
w2v.plot_words(words, text_size=7, fig_size=(12,6))
```
