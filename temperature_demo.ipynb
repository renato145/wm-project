```python
size = 100

x = np.arange(size)
y = np.random.random(size)

plt.figure(figsize=(10,7))
plt.bar(x,y)
```

```python
temperature = 0.25

preds = np.asarray(y).astype('float64')
preds = np.log(preds) / temperature
exp_preds = np.exp(preds)
preds = exp_preds / np.sum(exp_preds)

plt.figure(figsize=(10,7))
plt.bar(x,preds)
```
