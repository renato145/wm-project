```python
words = ['madre', 'padre', 'hombre', 'mujer', 'familia', 'pueblo', 'país', 'amigo', 'amigos',
         'hijo', 'hija', 'hermano', 'hermana', 'dios', 'santo', 'río', 'jesús', 'maría',
         'espíritu', 'guerra', 'engaño', 'pecado', 'joven', 'viejo', 'edad', 'año', 'años',
         'día', 'días', 'mes', 'agua', 'ríos', 'mar', 'sol', 'luna', 'noche', 'pecados']
w2v.plot_words(words, text_size=10, fig_size=(12,6))
```
